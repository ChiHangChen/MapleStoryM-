# Mission Go

```python
import pyscreenshot as ImageGrab
import win32gui
import numpy as np
import keyboard
import time
import cv2
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
def mission_go():
    can_go_mission=True
    keyboard.press_and_release('h')
    phoenix_time=time.time()
    while True:
        clear_output(wait=True)
        print(time.localtime()[3:6])
        im=ImageGrab.grab(bbox=(rect[0],rect[1]+35,rect[2],rect[3]))
        im=np.array(im)
        im=cv2.resize(im[:,:,:],(512,288))
        im=np.expand_dims(im,0)
        print(plt.imshow(im[0]))
        whichmax=np.argmax(model.predict(im)[0])
        if whichmax==0:
            print("chat")
            keyboard.press_and_release('n')
            keyboard.press_and_release('b')
        elif whichmax==1 and can_go_mission:
            print("ClickZ or Idle or NoQuest")
            can_go_mission=False
            im2=im[:,100:150,0:100,:]
            whichmax2=np.argmax(model2.predict(im2)[0])
            if whichmax2==0:
                print("ClickZ")
                keyboard.press_and_release('z')
                time.sleep(3)
            elif whichmax2==1:
                print("NoQuest")
                keyboard.press_and_release('i')
                time.sleep(3)
                keyboard.press_and_release('t')
                time.sleep(3)
                keyboard.press_and_release('q')
                time.sleep(3)
                keyboard.press_and_release('n')
                time.sleep(5)
        elif whichmax==2:
            print("Mission Complete")
            can_go_mission=True
            keyboard.press_and_release('c')
            time.sleep(3)
        elif whichmax==3 and can_go_mission:
            print("In Mulung")
        else:
            print("Other")
            time.sleep(5)
        if time.time()-phoenix_time > 10*60+0.1:
            time.sleep(0.5)
            keyboard.press_and_release('h')
            phoenix_time=time.time()
            time.sleep(0.5)
        time.sleep(0.5)
print("Done")
```

```python
for i in [5,4,3,2,1]:
    print(i)
    time.sleep(1)
    clear_output(wait=True)
if __name__=="__main__":
    rect=win32gui.GetWindowRect(win32gui.FindWindow(None, "MSM"))
    mission_go()
```

# MuLung Go

```python
def mulung_go():
    can_go_mulung=True
    keyboard.press_and_release('h')
    phoenix_time=time.time()
    
    #Use bow
    keyboard.press_and_release('l')
    bow_time=time.time()
    use_second_bow=1
    time.sleep(1)
    
    while True:
        clear_output(wait=True)
        print(time.localtime()[3:6])
        im=ImageGrab.grab(bbox=(rect[0],rect[1]+35,rect[2],rect[3]))
        im=np.array(im)
        im=cv2.resize(im[:,:,:],(512,288))
        im=np.expand_dims(im,0)
        print(plt.imshow(im[0]))
        whichmax=np.argmax(model.predict(im)[0])
        if whichmax==0:
            print("chat")
            # Chat
        elif whichmax==1:
            print("ClickZ or Idle")
            # Click Z or Idle or in mission
            keyboard.press_and_release('r')
            time.sleep(2)
            keyboard.press_and_release('h')
            time.sleep(2)
            keyboard.press_and_release('n')
            time.sleep(4)
            keyboard.press_and_release('e')
            time.sleep(5)
        elif whichmax==2:
            print("Mission Complete")
#             MissionComplete
            keyboard.press_and_release('c')
            time.sleep(3)
        elif whichmax==3:
            print("InMuLung")
            time.sleep(5)
        else:
            print("Other")
            time.sleep(5)
    
    
        if time.time()-phoenix_time > 10*60+0.1:
            keyboard.press_and_release('h')
            phoenix_time=time.time()
            time.sleep(1)
            
        if use_second_bow==1 and time.time() - bow_time > 10.1:
            keyboard.press_and_release('l')
            use_second_bow=0
            time.sleep(1)

        if time.time()-bow_time > 30+0.1 :
            keyboard.press_and_release('l')
            bow_time=time.time()
            use_second_bow=1
            time.sleep(1)
        
        time.sleep(0.5)
print("Done")
```
